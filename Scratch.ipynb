## What need to build ##
A simple two player Tic Tac Toe game that runs in the command line.

## Requirements
Two player gaming system
## Show Board 
show 3 x 3 grid on the screen
Show numbers 1-9 so players know which position to pick
Update the board after every move
## . Choose Symbol
Player 1 picks X or O
Player 2 automatically gets the other
Player 1 → picks X
Player 2 → gets O automatically

## Check for Winner

After every move check if someone won
There are 8 ways to win:
Rows:      1-2-3 / 4-5-6 / 7-8-9
Columns:   1-4-7 / 2-5-8 / 3-6-9
Diagonals: 1-5-9 / 3-5-7
If someone wins → announce winner and stop game

## Check for Draw

If all 9 cells are filled and nobody won
Announce it's a draw and stop game

## End the Game

Game stops when:
Someone wins ✅
It's a draw 🤝
Print final message and exit

## Play again
  Reset the game to start the playing game from scratch

## Simple game flow
START
  ↓
Show Board
  ↓
Player 1 picks X or O
  ↓
LOOP:
  Show Board
  Current player picks position
  Place symbol on board
  Anyone won? → YES → "Player X wins!" → STOP
  Board full?  → YES → "It's a draw!"  → STOP
  Switch to other player
  ↓
END




## Pseudocode for Tic Tac Toe Game ##

## Step 1: Write Pseudocode First

START PROGRAM

  1. Create empty board [1,2,3,4,5,6,7,8,9]
  2. Ask Player 1 to choose X or O
  3. Give Player 2 the other symbol

  REPEAT (while game is not over):
    4. Show the board
    5. Ask current player to pick a position
    6. Place their symbol on the board
    7. Check if they won → YES → announce winner → STOP
    8. Check if board is full → YES → announce draw → STOP
    9. Switch to other player

  Ask: Play again? → YES → go back to step 1
  
END PROGRAM
 
## Step 2: Break Into Functions

FUNCTION display_board(board):
    print the 3x3 grid
    show numbers for empty cells
    show X or O for filled cells

FUNCTION choose_symbol():
    ask player 1 to pick X or O
    give player 2 the other one
    return both symbols

FUNCTION get_move(board, player):
    ask player to pick position 1-9
    check if position is already taken
    if taken → ask again
    return valid position

FUNCTION check_winner(board, symbol):
    check all 8 winning combinations
    if any combination matches → return True
    else → return False

FUNCTION check_draw(board):
    check if all 9 cells are filled
    if full and no winner → return True
    else → return False

FUNCTION reset_board():
    clear the board back to [1..9]
    reset turn counter to 0
    return fresh board

## Step 3: Write the Main Structure

MAIN PROGRAM:

  SET board = [1,2,3,4,5,6,7,8,9]
  SET turn = 0

  CALL choose_symbol()
    → get player1, player2 symbols

  LOOP forever:

    CALL display_board(board)

    IF turn is even:
      current_player = player1
    ELSE:
      current_player = player2

    CALL get_move(board, current_player)
      → get position from player
      → place symbol on board

    CALL check_winner(board, current_player)
      IF True:
        print "Player X wins!"
        BREAK out of loop

    CALL check_draw(board)
      IF True:
        print "It's a draw!"
        BREAK out of loop

    ADD 1 to turn counter

  ASK "Play again? (yes/no)"
    IF yes → CALL reset_board() → start loop again
    IF no  → print "Thanks for playing!" → END



```python
## STEP 1: Create the Board and Variables
# ==========================================
# A simple list of 9 spots for our board
while True:
    # Fresh board for a new game
    board = ['1', '2', '3', 
             '4', '5', '6', 
             '7', '8', '9']

    # We start with player 'X'
    current_player = 'X'

    print("Welcome to Tic-Tac-Toe!")

    # ==========================================
    # STEP 2: The Main Game Loop
    # ==========================================
    # This loop keeps running until someone wins or ties
    while True:
    # 1. Show the board to the player
        print("\n")
        print(" " + board[0] + " | " + board[1] + " | " + board[2] + " ")
        print("---|---|---")
        print(" " + board[3] + " | " + board[4] + " | " + board[5] + " ")
        print("---|---|---")
        print(" " + board[6] + " | " + board[7] + " | " + board[8] + " ")
        print("\n")

        # 2. Ask the player for their move
        choice = input("Player " + current_player + ", choose a spot (1-9): ")
    
        # 3. Check if the input is valid (must be 1 to 9)
        if choice not in ['1', '2', '3', '4', '5', '6', '7', '8', '9']:
            print("Invalid input! Please type a number from 1 to 9.")
            continue # This restarts the loop so they can try again

          
        # Convert their text choice into a list number (index)
        # Example: Spot '1' is actually index 0 in Python
        index = int(choice) - 1
    
        # 4. Check if that spot is already taken
        if board[index] == 'X' or board[index] == 'O':
            print("That spot is already taken! Try another one.")
            continue # This restarts the loop so they can try again
        
        # 5. Place the player's mark on the board
        board[index] = current_player

# ==========================================
        # STEP 3: Check for a Winner
        # ==========================================
        # We check all 8 possible winning combinations manually using simple IF statements
        is_winner = False
    
        # Check Rows
        if board[0] == board[1] == board[2]: is_winner = True
        if board[3] == board[4] == board[5]: is_winner = True
        if board[6] == board[7] == board[8]: is_winner = True
        # Check Columns
        if board[0] == board[3] == board[6]: is_winner = True
        if board[1] == board[4] == board[7]: is_winner = True
        if board[2] == board[5] == board[8]: is_winner = True
        # Check Diagonals
        if board[0] == board[4] == board[8]: is_winner = True
        if board[2] == board[4] == board[6]: is_winner = True
    
        if is_winner == True:
            # Show final board
            print("\n")
            print(" " + board[0] + " | " + board[1] + " | " + board[2] + " ")
            print("---|---|---")
            print(" " + board[3] + " | " + board[4] + " | " + board[5] + " ")
            print("---|---|---")
            print(" " + board[6] + " | " + board[7] + " | " + board[8] + " ")
            print("\n")
            print("Congratulations! Player " + current_player + " wins! 🎉")
            break # Ends the game
        
        # ==========================================
        # STEP 4: Check for a Tie
        # ==========================================
        # If there are no numbers left on the board, it means it's full
        is_tie = True
        for spot in board:
            if spot != 'X' and spot != 'O':
                is_tie = False # Found a number, so game can continue
            
        if is_tie == True:
            print("It's a tie! 👔")
            break # Ends the game
        
        # ==========================================
        # STEP 5: Switch Players
        # ==========================================
        if current_player == 'X':
            current_player = 'O'
        else:
            current_player = 'X'

# ==========================================
# POST-MATCH: This runs ONLY when a match ends
# ==========================================
print("-" * 30)
play_again = input("Do you want to play another game? (yes/no): ")
    
if play_again.lower() != "yes" and play_again.lower() != "y":
    print("Thanks for playing! Goodbye! 👋")
    break  # Stops the outer loop entirely
```
